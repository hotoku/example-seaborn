seabornのlegendで特定の属性だけを表示する

```python
x = [1,2,3]
prop1 = ["a", "b"]
prop2 = ["c", "d"]
```

```python
import pandas as pd
import itertools
tups = list(itertools.product(x, prop1, prop2))
df = (
    pd
    .DataFrame(tups, columns=["x", "prop1", "prop2"])
    .assign(y = range(len(tups)))
)
df
```

```python
import seaborn as sns
ax = sns.lineplot(data=df, x="x", y="y", hue="prop1", style="prop2")
```

ここで、prop2のlegendだけ表示したい

```python
from matplotlib.axes import Axes

def filter_legend(ax: Axes, left: list[str], removed: list[str]):
    handles, labels = ax.get_legend_handles_labels()
    h2, l2 = [], []
    flag = True
    for h, l in zip(handles, labels):
        if l in removed:
            flag = False
        if l in left:
            flag = True
        if not flag:
            continue
        h2.append(h)
        l2.append(l)
    ax.legend(h2, l2)
            
```

```python
import seaborn as sns
ax = sns.lineplot(data=df, x="x", y="y", hue="prop1", style="prop2")
filter_legend(ax, ["prop2"], ["prop1"])
```

```python
filter_legend(ax, ["prop1"], ["prop2"])
```
